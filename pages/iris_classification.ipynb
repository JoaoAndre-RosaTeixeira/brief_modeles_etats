```python
# Hors sujet trouver si une fleur est une Iris
```

```python
# choix de deux variables
X = iris.data[:, :2] # Utiliser les deux premiers colonnes afin d'avoir un problème de classification binaire.&nbsp;
y = (iris.target != 0) * 1 # re-étiquetage des fleurs
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(X[y == 0][:, 0], X[y == 0][:, 1], color='g', label='0')
plt.scatter(X[y == 1][:, 0], X[y == 1][:, 1], color='y', label='1')
plt.legend();
plt.show()
```

```python
# on remarque par ces résultat que les Iris sont dans la partie jaune et ce qui n'est pas conciderer comme Iris dans la partie verte.
```

```python
model = LogisticRegression(C=1e20) # construction d'un objet de Régression logistique
model.fit(X, y) # Entrainement du modèle
```

```python
# on implemente les nouvelels fleurs à evaluer.
Iries_To_Predict = [
    [5.5, 2.5],
    [7, 3],
    [3,2],
    [5,3]
]
```

```python
model.predict(Iries_To_Predict) 
```

```python
# on remarque par ces résultat que les 2 premiere fleurs sont bien des Iris tant dis que les deux dernires n'en sont pas.
```
